### Adequação do Dicionário de Referência

A adequação do dicionário de referência foi uma das partes mais importantes da transformação dos arquivos da camada bronze para a camada prata. Afinal, é ele quem possibilita a tradução dos códigos da pesquisa, tornando-a legível para seres humanos.

**Valores vazios**

Ao importar o dicionário, que se encontra no formato .xlsx, para o Pandas, um problema que encontramos foi que as células mescladas do Excel se tornaram todas vazias, com exceção da primeira do conjunto. Arrumar isso foi o primeiro passo da adequação do dicionário de referência, o que fizemos por meio da aplicação da função .ffill() nas colunas que possuíam células mescladas. Além disso, também apagamos as linhas correspondentes aos cabeçalhos de seção do dicionário, que não servem para a tradução de códigos.

<img src='https://drive.google.com/uc?id=1uxhob2QoV21_ZOoEK78712mnAXPSG4s6' width=700>

**Coluna de códigos**

Outro problema que encontramos foi a existência, na coluna de códigos, de valores que não correspondiam a códigos, mas expressavam o limite de valores possível para aquela varíavel. Isso acontecia principalmente em variáveis numéricas, com relação às quais não faz sentido a existência de códigos. Podemos observar três casos desses a seguir:
![Não-códigos](https://drive.google.com/uc?id=1e-OTtUVodnITv1I2znJJCekL3B5YJOvw)
Como esses valores não representam códigos de verdade, excluímo-los apagando todos os valores da coluna que continham letras.

**Coluna de significados**

Nesta coluna, um dos problemas encontrados foi a existência de significados que não estavam associados a nenhum código, como "Não aplicável", que serve como significado para qualquer célula vazia da pesquisa. Além disso, também houve a questão do significado "Ignorado", que é essencialmente a mesma coisa, mas com um código associado a si. O problema disso é que, por causa de sua existência, colunas que eram de fato booleanas pareciam ser categóricas na prática. Observe esses casos a seguir:
![significados a excluir](https://drive.google.com/uc?id=1mNX0XgYJYh8eYXBgCQ6NidQ-cj_ePhUG)
Com relação ao significado "Não aplicável", bastou eliminá-lo (e suas variações) do dicionário. Já com relação a "Ignorado", porém, não podíamos simplesmente excluí-lo, ou seus códigos ficariam sem significado no momento de traduzir a pesquisa. A abordagem adotada foi a de deixá-lo como estava até a tradução da pesquisa e, uma vez traduzida, tornar vazia qualquer célula que tivesse o valor "Ignorado".

Obs.: o mesmo problema de valores associados a variáveis númericas que descrevemos com relação à coluna de códigos existia na coluna de significados. Simplesmente tornamos vazios esses valores.

**Variáveis desnecessárias**

Observando o dicionário de referência, identificamos algumas variáveis desnecessárias, pois reduntantes. Se pegarmos a variável "CAPITAL", por exemplo, podemos determiná-la completamente com base na variável "UF". A existência de variáveis para o dia, o mês e o ano de nascimento dos moradores também nos pareceu dispensável, uma vez que continham muitos valores vazios na pesquisa e podiam ser substituídas pela idade dos moradores. Uma lista completa das variáveis descartadas e das razões para isso é apresentada a seguir:

- CAPITAL: completamente redundante com a coluna UF e inteiramente determinada por ela.

- V1016-Número da entrevista no domicílio: nos é completamente irrelevante saber se essa é a primeira, a terceira ou a quinta entrevista feita com o mesmo domicílio. No máximo, isso importaria para ver a progressão da situação, mas isso já é possível avaliar com as colunas referentes à data de realização da pesquisa.

- V1030-Projeção da população: importante apenas para aqueles que querem fazer uma meta-análise dos pesos aplicados. Não vamos tão a fundo assim e, para nós, adotar os pesos já é suficiente.

- posest-Domínios de Projeção: descartável, uma vez que é a combinação de outras colunas.

- Dia, mẽs e ano de nascimento: são redundantes com a coluna de idade do morador. Poderíamos até adotar a data de nascimento como referência mas, na pesquisa, há várias entradas sem uma data de nascimento, mas com a idade do morador. É melhor, portanto, usarmos esta última.

- C007E1 e C007E2: totalmente reduntantes com a coluna C007E.

- C010 e C011A: colunas binárias que apenas indicam se a pergunta foi respondida. Para termos a mesma informação, basta vermos se as colunas que seguem cada uma dessas perguntas estão preenchidas ou não.

- C0101, C0102, C011A1, C011A2: na prática, também funcionam como colunas binárias. Podemos compensar sua exclusão da mesma forma que compensamos as colunas C010 e C011A.

- F006 e F0061: nos informam sobre qual foi o morador que respondeu à pesquisa.

Obviamente, o descarte dessas variáveis do dicionário foi acompanhado do descarte das mesmas colunas nos microdados. Também o foi a criação de uma nova coluna, C005_custom, que somou o tempo de afastamento do trabalho em meses expresso pelas colunas C0051, C0052 e C0053.

**Aprimoramento das descrições**

A coluna que continha as descrições das variáveis foi a coluna utilizada para nomear cada coluna dos microdados. Como ela continha valores muito grandes, porém, inadequados para o nome de uma coluna, utilizamos um dicionário do python para associar cada variável a uma nova descrição, mais adequada para nomear colunas. Aliás, aproveitamos a situação para adicionar prefixos que nos ajudariam mais à frente, na normalização da pesquisa em tabelas-dimensão e tabelas-fato. Descrições como "Tem diagnóstico de doença respiratória crônica ou doença de pulmão?", por exemplo, foram alteradas para "doenças_crônicas_doença_respiratória". O dicionário com todas as variáveis e descrições associadas pode ser encontrado no código disponibilizado ao final desta seção.

### Tradução e Conformação da Pequisa

Com o dicionário pronto, pudemos então passar às transformações dos microdados. Em primeiro lugar, foi necessário importá-los via Spark utilizando o esquema do último mês, ou o Spark não daria conta das novas colunas que foram adicionadas nos meses posteriores.

**Ajustes iniciais**

Com os dados importados, o primeiro passo foi descartar as colunas desnecessárias que já mencionamos e criar a coluna C005_custom, também já mencionada. Além disso, criamos uma coluna de datas para facilitar a análise temporal, colunas com ids numéricos para os domicílios e seus moradores, o que agiliza as joins, e aplicamos os deflatores encontrados no arquivo "Deflator_PNAD_COVID_2020_11.csv". Também foi criada uma coluna "estrato_geográfico" a partir dos quatro primeiros números de "Estrato" para possibilitar a análise geográfica mais precisa dos dados.

**Tradução da pesquisa**

Feitos esses ajustes iniciais, nossa pesquisa estava finalmente pronta para ser traduzida. A lógica da tradução pode ser analisada detalhadamente no código apresentado no fim desta seção, mas a ideia geral é que cada código na coluna de variáveis deveria ser substituído por uma descrição que se tornaria o nome da coluna e cada código da coluna de códigos deveria ser substituído por um significado que se tornaria o valor das células. Ou seja, a tradução da pesquisa com base no dicionário fez as seguintes transformações: variável -> descrição (para colunas); código -> significado (para células).

**Conformação da pesquisa**
Com a pesquisa traduzida, pudemos então passar à validação dos valores e do tipo de dados de cada coluna. Foi neste momento, por exemplo, que excluímos o valor "Ignorado" de todas as células, o que nos permitiu transformar diversas colunas em colunas booleanas. Também arredondamos o valor dos alugueis para transformá-los em uma coluna do tipo int e transformamos colunas com valores até 127 em ByteType.

Além de diminuir o tamanho de armazenamento do arquivo, essas transformações foram aplicadas visando melhorar a performance do Spark para processá-los.

**Pesquisa ajustada**

A pesquisa traduzida e ajustada foi salva como arquivo .parquet na camada prata para normalização futura. Embora ela ainda vá ser normalizada, nenhum de seus valores será transformado daqui para frente, o que faz com que esse arquivo seja um bom ponto de validação e debug.

### Script AWS Glue Utilizado para Realizar as Transformações Mencionadas

```python
# --- Imports ---

import functools
import os
import re
import sys
from collections import Counter
from datetime import date, timedelta

import boto3
import numpy as np
import pandas as pd
import geopandas as gpd

from pyspark.context import SparkContext
from pyspark.sql.functions import *
from pyspark.sql.types import *

from awsglue.transforms import *
from awsglue.utils import getResolvedOptions
from awsglue.context import GlueContext
from awsglue.job import Job

# --- Parâmetros do Job e Inicialização do Contexto ---

args = getResolvedOptions(sys.argv, ['JOB_NAME', 'S3_BUCKET'])
bucket_name = args['S3_BUCKET']

sc = SparkContext.getOrCreate()
glueContext = GlueContext(sc)

spark = glueContext.spark_session

job = Job(glueContext)
job.init(args['JOB_NAME'], args)

s3_client = boto3.client('s3')

# --- Definição das funções ---

def microdados_import(bucket, prefix):
    response = s3_client.list_objects_v2(Bucket=bucket, Prefix=prefix)

    keys = [
        obj['Key'] for obj in response.get('Contents', [])
        if re.search(r'PNAD_COVID_(\d+)\.csv', obj['Key'])
    ]

    dfs = []
    for key in keys:
        s3_file_path = f"s3://{bucket}/{key}"
        df = spark.read.option("header", True).option("inferSchema", True).csv(s3_file_path)
        dfs.append(df)

    if not dfs:
        raise ValueError(f"Nenhum arquivo encontrado no bucket {bucket} com o prefixo {prefix}.")

    df = functools.reduce(lambda a, b: a.unionByName(b, allowMissingColumns=True), dfs)

    return df


def create_c005_custom(df):
    """Cria a coluna C005_custom, que sintetiza as colunas C0051, C0052 e C0053."""
    df = df.withColumnRenamed("C005", "C005_custom")
    df = df.withColumn("C005_custom",
                       coalesce(col('C0051'), lit(0)) +
                       coalesce(col('C0052'), lit(0)) +
                       12 * coalesce(col('C0053'), lit(0)))
    df = df.withColumn("C005_custom", when(col("C005_custom") == 0, None).otherwise(col("C005_custom")))
    df = df.drop('C0051', 'C0052', 'C0053')

    return df


def create_date_col(df, col_ano, col_mes, col_semana):
    def get_date(year, month, week):
        # Find the first day of the month
        first_day = date(int(year), int(month), 1)
        # Find the first Monday of the month
        first_monday = first_day + timedelta(days=(7 - first_day.weekday()) % 7)
        # Go to the Monday of the desired week
        week_monday = first_monday + timedelta(weeks=int(week) - 1)
        # Add 2 days to get to Wednesday
        return week_monday + timedelta(days=2)

    get_date_udf = udf(get_date, DateType())
    df = df.withColumn('data', get_date_udf(col(col_ano), col(col_mes), col(col_semana)))

    return df


def create_hash_id(df, new_col, key_cols):
    """Concatena colunas e cria um hash_id para aumento de performance em joins."""
    parts = []
    for i, c in enumerate(key_cols):
        parts.append(col(c).cast("string"))
        if i < len(key_cols) - 1:
            parts.append(lit("_"))

    return df.withColumn(new_col, hash(concat(*parts)).cast("long"))


def apply_deflatores(df, deflatores_path, esquema_deflatores):
    deflatores = (spark.read.option("header", "true").option("sep", ";")
                  .schema(esquema_deflatores).csv(deflatores_path))
    deflatores = deflatores.withColumnsRenamed(
        {'Ano': 'Ano_2',
         'MES': 'Mês',
         'UF': 'UF_2'}
    )

    for c in ["Habitual", "Efetivo", "CO3"]:
        deflatores = deflatores.withColumn(c, regexp_replace(col(c), ",", ".").cast("double"))

    df = df.join(deflatores,
                 (df["Ano"] == deflatores["Ano_2"]) &
                 (df["V1013"] == deflatores["Mês"]) &
                 (df["UF"] == deflatores["UF_2"]),
                 how="left")

    habitual_cols = ['C01012', 'C01022']
    efetivo_cols = ['C011A12', 'C011A22', 'D0013', 'D0023', 'D0033', 'D0043', 'D0053', 'D0063', 'D0073', 'F0021']

    for valor_habitual in habitual_cols:
        df = df.withColumn(valor_habitual, round(col(valor_habitual) * col("Habitual"), 2))

    for valor_efetivo in efetivo_cols:
        df = df.withColumn(valor_efetivo, round(col(valor_efetivo) * col("Efetivo"), 2))

    df = df.drop('Ano_2', 'Mês', 'UF_2', 'Habitual', 'Efetivo', 'CO3')

    return df


def download_shapefile(bucket, prefix, local_dir):
    response = s3_client.list_objects_v2(Bucket=bucket, Prefix=prefix)
    for obj in response.get('Contents', []):
        file_key = obj['Key']
        local_file_path = os.path.join(local_dir, os.path.basename(file_key))
        s3_client.download_file(bucket, file_key, local_file_path)


def consolidate_geometria_estratos(df, estratos_mun_path, mun_geometry_path):
    estratos_ibge = pd.read_csv(estratos_mun_path, sep=';')
    municipios = gpd.read_file(mun_geometry_path)

    municipios = municipios[['CD_MUN', 'geometry']]
    estratos_ibge['Código do Município'] = estratos_ibge['Código do Município'].astype(str)
    estratos_ibge = municipios.merge(estratos_ibge, left_on='CD_MUN', right_on='Código do Município')

    estratos_geo = estratos_ibge.dissolve(by='Código do estrato', as_index=False)[['Código do estrato', 'geometry']]
    mun_agg = estratos_ibge.groupby('Código do estrato').agg(
        nome_estrato = ('Nome abreviado do estrato', 'first'),
        municipios = ('Nome do Município', list)).reset_index()

    estratos = estratos_geo.merge(mun_agg, on='Código do estrato', how='left')

    estratos = estratos[['Código do estrato', 'nome_estrato', 'municipios', 'geometry']]
    estratos.columns = ['estrato_geográfico', 'nome_estrato', 'municípios', 'geometry']

    df = df.withColumn('estrato_geográfico',
                       substring(df['Estrato'].cast('string'), 1, 4)
                       .cast('int'))
    df.drop('Estrato')

    return df, estratos


def adjust_dic_pnad(dic_path, desc_rename_dict):
    dic_pnad = pd.read_excel(dic_path,
                             skiprows=4,
                             header=None,
                             usecols="B:F",
                             names=['variável', 'questão', 'descrição', 'código', 'significado'],
                             dtype='str')

    dic_pnad = dic_pnad.dropna(how='all')
    dic_pnad[['variável', 'questão', 'descrição']] = dic_pnad[['variável', 'questão', 'descrição']].ffill()
    dic_pnad.loc[dic_pnad['código'].str.contains(r'[a-zA-Z]', na=False), 'código'] = np.nan
    dic_pnad = dic_pnad[~dic_pnad['significado'].isin(['Não aplicável', 'não aplicável', 'Não aplicado'])]
    dic_pnad.loc[dic_pnad['código'].isna(), 'significado'] = np.nan
    dic_pnad = dic_pnad.replace(r'\n', ' ', regex=True).reset_index(drop=True)

    idx_c005 = dic_pnad[dic_pnad['variável'] == 'C005'].index[0]
    dic_pnad.iloc[idx_c005] = ['C005_custom', 'C5', 'Há quanto tempo está afastado desse trabalho?', np.nan, np.nan]

    cols_to_drop = ['CAPITAL', 'V1016', 'V1030', 'posest', 'A001B1', 'A001B2', 'A001B3', 'C005', 'C0051', 'C0052',
                    'C0053', 'C007E1', 'C007E2', 'C010', 'C0101', 'C0102', 'C011A', 'C011A1', 'C011A2',
                    'D0011', 'D0021', 'D0031', 'D0041', 'D0051', 'D0061', 'D0071','F006', 'F0061']
    dic_pnad = dic_pnad[~dic_pnad['variável'].isin(cols_to_drop)].reset_index(drop=True)

    for var, desc in desc_rename_dict.items():
        dic_pnad.loc[dic_pnad['variável'] == var, 'descrição'] = desc

    filtro = dic_pnad['descrição'] == 'area_tipo'
    dic_pnad.loc[filtro, 'significado'] = dic_pnad.loc[filtro, 'significado'].str.replace(r'\s*\(.*?\)', '', regex=True)

    for col in dic_pnad.columns:
        dic_pnad[col] = dic_pnad[col].str.strip()

    dic_pnad['código'] = dic_pnad['código'].astype('Int64')

    return dic_pnad


def collect_dict(dic_pnad):
    col_rename_map = {}
    col_mappings = {}

    for _, row in dic_pnad.iterrows():
        col_rename_map[row['variável']] = row['descrição']
        if pd.notna(row['código']):
            col_mappings.setdefault(row['variável'], {})[int(row['código'])] = row['significado']

    return col_rename_map, col_mappings


def translate_df(df, dic_pnad):
    col_rename_map, col_mappings = collect_dict(dic_pnad)

    for col_name, mapping_dict in col_mappings.items():
        if col_name in df.columns:
            mapping_expr = create_map([lit(x) for pair in mapping_dict.items() for x in pair])
            df = df.withColumn(
                col_name,
                coalesce(
                    mapping_expr[col(col_name).cast("string")],
                    col(col_name).cast("string")
                )
            )

    renamed = [col_rename_map.get(c, c) for c in df.columns]
    name_counts = Counter(renamed)
    final_names = []
    for original, new_name in zip(df.columns, renamed):
        if name_counts[new_name] > 1:
            final_names.append(f"{new_name} ({original})")
        else:
            final_names.append(new_name)

    return df.toDF(*final_names)


def ajust_col_dtypes(df, bool_cols, small_nums_cols):
    """Esta função visa ajustar os tipos de dado de cada coluna agora que seus valores foram
    traduzidos com base no dicionário. Para isso, porém, ainda são implementadas mais algumas
    alterações em determinadas colunas."""
    df = df.replace('Ignorado', None)

    for c in bool_cols:
        df = df.withColumn(
            c,
            when(col(c) == 'Sim', True)
            .when(col(c) == 'Não', False)
            .otherwise(None)
            .cast(BooleanType())
        )

    df = df.withColumn(
        'escola_pública',
        when(df['escola_pública'] == 'Pública', True)
        .when(df['escola_pública'] == 'Privada', False)
        .otherwise(None)
        .cast(BooleanType())
        )

    df = df.withColumn(
        'rendimento_recebia_somente_em_benefícios',
        when(df['rendimento_recebia_somente_em_benefícios'] == 'Em benefícios', True)
        .otherwise(None)
        .cast(BooleanType())
        )

    df = df.withColumn(
        'rendimento_era_não_remunerado',
        when(df['rendimento_era_não_remunerado'] == 'Não remunerado', True)
        .otherwise(None)
        .cast(BooleanType())
    )

    df = df.withColumn(
        'domicílio_valor_aluguel',
        round(col('domicílio_valor_aluguel')).cast(IntegerType())
    )

    for c in small_nums_cols:
        df = df.withColumn(
            c,
            col(c).cast(ByteType())
        )

    return df

# --- Variáveis auxiliares (listas e dicionários) ---

cols_to_drop = ['CAPITAL', 'V1016', 'V1030', 'posest',
                'A001B1', 'A001B2', 'A001B3',
                'C007E1', 'C007E2', 'C010', 'C0101', 'C0102',
                'C011A', 'C011A1', 'C011A2',
                'D0011', 'D0021', 'D0031', 'D0041', 'D0051', 'D0061', 'D0071',
                'F006', 'F0061']

esquema_deflatores = StructType([
    StructField("Ano", IntegerType()),
    StructField("MES", IntegerType()),
    StructField("UF", IntegerType()),
    StructField("Habitual", StringType()),
    StructField("Efetivo", StringType()),
    StructField("CO3", StringType())
])

desc_rename_dict = {
    'Ano': 'ano',
    'UF': 'unidade_federativa',
    'RM_RIDE': 'região_administrativa',
    'V1008': 'domicílio_número',
    'V1012': 'semana_mês',
    'V1013': 'mês',
    'Estrato': 'estrato',
    'V1022': 'domicílio_zona',
    'V1023': 'area_tipo',
    'V1031': 'domicílio_peso',
    'V1032': 'morador_peso',
    'A001': 'morador_número',
    'A001A': 'morador_parentesco',
    'A002': 'morador_idade',
    'A003': 'morador_sexo',
    'A004': 'morador_raça',
    'A005': 'morador_escolaridade',
    'A006': 'morador_estuda',
    'A006A': 'escola_pública',
    'A006B': 'escola_aulas_presenciais',
    'A007': 'escola_lição_de_casa',
    'A007A': 'escola_lição_de_casa_porque_não_fez',
    'A008': 'escola_dedicação',
    'A009': 'escola_dedicação_tempo',
    'B0011': 'sintomas_febre',
    'B0012': 'sintomas_tosse',
    'B0013': 'sintomas_dor_de_garganta',
    'B0014': 'sintomas_dificuldade_para_respirar',
    'B0015': 'sintomas_dor_de_cabeça',
    'B0016': 'sintomas_dor_no_peito',
    'B0017': 'sintomas_náusea',
    'B0018': 'sintomas_nariz_escorrendo',
    'B0019': 'sintomas_fadiga',
    'B00110': 'sintomas_dor_nos_olhos',
    'B00111': 'sintomas_perda_de_olfato',
    'B00112': 'sintomas_dor_muscular',
    'B00113': 'sintomas_diarreia',
    'B002': 'providencias_procurou_atendimento',
    'B0031': 'providencias_ficou_em_casa',
    'B0032': 'providencias_ligou_para_médico',
    'B0033': 'providencias_automedicação',
    'B0034': 'providencias_tomou_remédio_orientação_médica',
    'B0035': 'providencias_visita_domiciliar_SUS',
    'B0036': 'providencias_visita_domiciliar_médico_particular',
    'B0037': 'providencias_outra',
    'B0041': 'local_atendimento_posto_de_saúde',
    'B0042': 'local_atendimento_pronto_socorro_SUS',
    'B0043': 'local_atendimento_hospital_SUS',
    'B0044': 'local_atendimento_consultório_privado',
    'B0045': 'local_atendimento_pronto_socorro_privado',
    'B0046': 'local_atendimento_hospital_privado',
    'B005': 'internação_internou',
    'B006': 'internação_entubou',
    'B007': 'tem_plano_de_saúde',
    'B008': 'exames_fez_algum',
    'B009A': 'exames_fez_SWAB',
    'B009B': 'exames_resultado_SWAB',
    'B009C': 'exames_fez_furo_no_dedo',
    'B009D': 'exames_resultado_furo_no_dedo',
    'B009E': 'exames_fez_punção_venosa',
    'B009F': 'exames_resultado_punção_venosa',
    'B0101': 'doenças_crônicas_diabetes',
    'B0102': 'doenças_crônicas_hipertensão',
    'B0103': 'doenças_crônicas_doença_respiratória',
    'B0104': 'doenças_crônicas_doenças_cardíacas',
    'B0105': 'doenças_crônicas_depressão',
    'B0106': 'doenças_crônicas_câncer',
    'B011': 'prevenção_restringiu_contato',
    'C001': 'trabalho_trabalhou_ou_fez_bico',
    'C002': 'trabalho_estava_afastado',
    'C003': 'trabalho_afastamento_motivo',
    'C004': 'trabalho_foi_remunerado_enquanto_afastado',
    'C005_custom': 'trabalho_meses_afastado',
    'C006': 'trabalho_tem_mais_de_um',
    'C007': 'trabalho_tipo',
    'C007A': 'trabalho_esfera',
    'C007B': 'trabalho_clt_ou_funcionário_público',
    'C007C': 'trabalho_função',
    'C007D': 'trabalho_ramo_empresa',
    'C007E': 'trabalho_tamanho_empresa',
    'C007F': 'trabalho_contrato_suspenso',
    'C008': 'trabalho_quantas_horas_trabalhava',
    'C009': 'trabalho_quantas_horas_trabalhou',
    'C009A': 'trabalho_queria_ter_trabalhado_mais',
    'C01011': 'rendimento_faixa_normalmente_recebida_dinheiro',
    'C01012': 'rendimento_valor_normalmente_recebido_dinheiro',
    'C01021': 'rendimento_faixa_normalmente_recebida_produtos_e_mercadorias',
    'C01022': 'rendimento_valor_normalmente_recebido_produtos_e_mercadorias',
    'C0103': 'rendimento_recebia_somente_em_benefícios',
    'C0104': 'rendimento_era_não_remunerado',
    'C011A11': 'rendimento_faixa_efetivamente_recebida_dinheiro',
    'C011A12': 'rendimento_valor_efetivamente_recebido_dinheiro',
    'C011A21': 'rendimento_faixa_efetivamente_recebida_produtos_e_mercadorias',
    'C011A22': 'rendimento_valor_efetivamente_recebido_produtos_e_mercadorias',
    'C012': 'trabalho_trabalhou_presencialmente',
    'C013': 'trabalho_trabalhou_remotamente',
    'C014': 'trabalho_contribui_INSS',
    'C015': 'trabalho_procurou_emprego',
    'C016': 'trabalho_porque_não_procurou_emprego',
    'C017A': 'trabalho_gostaria_de_ter_trabalhado',
    'D0013': 'renda_domiciliar_aposentadoria_pensão',
    'D0023': 'renda_domiciliar_pensão_alimentícia_doação_mesada',
    'D0033': 'renda_domiciliar_bolsa_família',
    'D0043': 'renda_domiciliar_benefício_assistencial_continuado',
    'D0053': 'renda_domiciliar_auxílio_emergencial_coronavirus',
    'D0063': 'renda_domiciliar_seguro_desemprego',
    'D0073': 'renda_domiciliar_outros_rendimentos',
    'E001': 'empréstimo_domicílio_alguém_pediu',
    'E0021': 'empréstimo_domicílio_banco',
    'E0022': 'empréstimo_domicílio_parente_ou_amigo',
    'E0023': 'empréstimo_domicílio_empregado_ou_patrão',
    'E0024': 'empréstimo_domicílio_outro_local_ou_pessoa',
    'F001': 'domicílio_situação',
    'F0021': 'domicílio_valor_aluguel',
    'F0022': 'domicílio_faixa_aluguel',
    'F002A1': 'profilaxia_domicílio_sabão_ou_detergente',
    'F002A2': 'profilaxia_domicílio_álcool_70%',
    'F002A3': 'profilaxia_domicílio_máscaras',
    'F002A4': 'profilaxia_domicílio_luvas_descartáveis',
    'F002A5': 'profilaxia_domicílio_água_sanitária_ou_desinfetante'
}

bool_cols = [
    'morador_estuda',
    'providencias_procurou_atendimento',
    'providencias_ficou_em_casa',
    'providencias_ligou_para_médico',
    'providencias_automedicação',
    'providencias_tomou_remédio_orientação_médica',
    'providencias_visita_domiciliar_SUS',
    'providencias_visita_domiciliar_médico_particular',
    'providencias_outra',
    'local_atendimento_posto_de_saúde',
    'local_atendimento_pronto_socorro_SUS',
    'local_atendimento_hospital_SUS',
    'local_atendimento_consultório_privado',
    'local_atendimento_pronto_socorro_privado',
    'local_atendimento_hospital_privado',
    'internação_entubou',
    'tem_plano_de_saúde',
    'exames_fez_algum',
    'exames_fez_SWAB',
    'exames_fez_furo_no_dedo',
    'exames_fez_punção_venosa',
    'doenças_crônicas_diabetes',
    'doenças_crônicas_hipertensão',
    'doenças_crônicas_doença_respiratória',
    'doenças_crônicas_doenças_cardíacas',
    'doenças_crônicas_depressão',
    'doenças_crônicas_câncer',
    'trabalho_trabalhou_ou_fez_bico',
    'trabalho_estava_afastado',
    'trabalho_tem_mais_de_um',
    'trabalho_contrato_suspenso',
    'trabalho_queria_ter_trabalhado_mais',
    'trabalho_trabalhou_presencialmente',
    'trabalho_trabalhou_remotamente',
    'trabalho_contribui_INSS',
    'trabalho_procurou_emprego',
    'trabalho_gostaria_de_ter_trabalhado',
    'empréstimo_domicílio_banco',
    'empréstimo_domicílio_parente_ou_amigo',
    'empréstimo_domicílio_empregado_ou_patrão',
    'empréstimo_domicílio_outro_local_ou_pessoa',
]

small_nums_cols = [
    'domicílio_número',
    'semana_mês',
    'mês',
    'morador_número',
    'morador_idade'
]

# --- Variáveis auxiliares (caminhos de arquivos) ---

pesquisa_prefix = 'Bronze Layer/Microdados/'
deflatores_path = 's3://tech-chall-3-pnad-covid/Bronze Layer/Deflator_PNAD_COVID_2020_11.csv'
estratos_mun_path = '/tmp/Municipios_por_Estratos.csv'
mun_geometry_prefix = 'Bronze Layer/BR_Municipios_2024/'
shapefile_dir = '/tmp/BR_Municipios/'
dic_path = '/tmp/dicionario.xls'

# --- Importação de arquivos para pasta temporária ---

os.makedirs(shapefile_dir, exist_ok=True)
download_shapefile(bucket_name, mun_geometry_prefix, shapefile_dir)
mun_geometry_path = os.path.join(shapefile_dir, 'BR_Municipios_2024.shp')
s3_client.download_file(bucket_name, 'Bronze Layer/Dicionario_PNAD_COVID_112020_20220621.xls', dic_path)
s3_client.download_file(bucket_name, 'Bronze Layer/Municipios_por_Estratos.csv', estratos_mun_path)

# --- Transformação dos dados ---

pesquisa = microdados_import(bucket_name, pesquisa_prefix)
pesquisa = pesquisa.drop(*cols_to_drop)
pesquisa = create_c005_custom(pesquisa)
pesquisa = create_date_col(pesquisa, 'Ano', 'V1013', 'V1012')
pesquisa = create_hash_id(pesquisa, "id_domicílio", ["UPA", "V1008"])
pesquisa = create_hash_id(pesquisa, "id_morador",   ["UPA", "V1008", "A001"])
pesquisa = apply_deflatores(pesquisa, deflatores_path, esquema_deflatores)
pesquisa, estratos = consolidate_geometria_estratos(pesquisa, estratos_mun_path, mun_geometry_path)
dic_pnad = adjust_dic_pnad(dic_path, desc_rename_dict)
pesquisa_trad = translate_df(pesquisa, dic_pnad)
pesquisa_ajust = ajust_col_dtypes(pesquisa_trad, bool_cols, small_nums_cols)

# --- Salvamento dos arquivos ---

dic_pnad.to_csv('/tmp/dicionario_ajustado.csv', index=False, sep=';')
s3_client.upload_file('/tmp/dicionario_ajustado.csv', bucket_name, 'Silver Layer/dicionario_ajustado.csv')

estratos.to_file("/tmp/estratos_geometry.geojson", driver="GeoJSON")
s3_client.upload_file('/tmp/estratos_geometry.geojson', bucket_name, 'Silver Layer/estratos_geometry.geojson')

with open('/tmp/esquema_pesquisa.json', 'w') as f:
    f.write(pesquisa_ajust.schema.json())
s3_client.upload_file('/tmp/esquema_pesquisa.json', bucket_name, 'Silver Layer/esquema_pesquisa.json')

s3_parquet_path = f"s3://{bucket_name}/Silver Layer/pesquisa_traduzida"
pesquisa_ajust.coalesce(1).write.mode("overwrite").parquet(s3_parquet_path)

job.commit()
```

## 2.3 Normalização dos Dados (camada prata -> camada ouro)

As transformações que descreveremos a partir daqui serão aquelas que aplicamos sobre os arquivos da Silver Layer e resultaram nos arquivos finais da Gold Layer.



Ao armazenarmos nossos dados, é importante aplicarmos alguns princípios da engenharia de dados. A normalização, por exemplo, evita a redundância e garante uma melhor gestão de nossos dados. Ao observarmos as colunas da pesquisa traduzida, vemos que faz sentido agrupar algumas delas em tabelas separadas, evitando assim repeti-los a todo momento e criando uma única fonte de verdade, o que também facilita a atualização dos dados, se necessária.

<br>

**Tabelas-Fato**

Ao separarmos os dados em diferentes tabelas, a principal questão que devemos considerar é se elas devem ser tabelas-fato ou tabelas-dimensão. Tabelas-fato são aquelas que agrupam os dados que efetivamente mensuram o fenômeno que estamos investigando e outros fatores relacionados a ele mas que, como ele, tendem a se alterar a cada observação. Pensando em nosso conjunto de dados, nossa principal tabela-fato será a dos sintomas e providências tomadas, uma vez que estamos investigando a pandemia de COVID. No entanto, outro fenômeno que podemos desvendar no questionário são os impactos da pandemia sobre a situação econômica dos cidadãos: alterações na quantidade de horas trabalhadas, nos valores recebidos, mudanças no regime de trabalho (presencial/remoto), etc.

As tabelas-fato que criaremos serão três: uma relativa à situação econômica dos cidadãos; outra relativa à disseminação da doença entre eles; e uma terceira, relativa às condições sanitárias e econômicas dos domicílios.

<br>

**Tabelas-Dimensão**

Diferentemente das tabelas-fato, as tabelas-dimensão não nos informam sobre o fenômeno em si que estamos investigando, mas sobre as entidades que participam de sua dinâmica. No caso da PNAD COVID, as principais entidades da pesquisa são os domicílios e seus moradores. Assim, as tabelas-dimensão devem ser construída pelo agrupamento de colunas que nos deem informações sobre esses entes. Diferentemente das tabelas-fato, porém, essas informações são muito mais perenes e tendem a não se alterar a cada nova observação (no caso, a cada nova aplicação do questionário). As tabelas-dimensão envolverão, portanto, informações como a UF e o estrato a que pertence o domicílio, bem como a profissão de seus moradores e seu grau de escolaridade.

<br>

Obs.: Como na seção anterior, disponibilizaremos, ao final da seção, o código utilizado para realizar as transformações que discutiremos aqui. Lá, o leitor pode encontrar a divisão exata das tabelas com suas respectivas colunas.

### dim_datas
A tabela de datas é um tipo especial de tabela-dimensão: diferentemente das demais, ela cresce a cada nova observação, ganhando um novo registro para cada dia em que os dados são colhidos. O que permanece constante nela são os campos relacionados a cada novo registro: uma data sempre estará associada ao mesmo mês, ano, semana, etc. A depender da tecnologia utilizada, uma tabela de datas nem é necessária: as bibliotecas do Python nos permitem facilmente fazer análises temporais utilizando somente a data das observações. Já outros serviços, como o Power BI, se beneficiam da construção de uma tabela de datas com campos como "Ano", "Mês", "Semana no Mês", "Semana no Ano", "Dia da Semana", etc.

Como utilizamos o Python para a análise de dados, não criamos uma tabela separada para datas. No entanto, achamos importante mencionar esse caso especial aqui. Caso fôssemos utilizar outra tecnologia, as colunas que seriam incluídas na tabela de datas seriam:
- ano
- mês
- semana_mês
- data

### dim_estratos

A tabela dim_estratos foi construída da junção de estratos_geometry.geojson, que criamos no pré-processamento dos dados, com as seguintes colunas da pesquisa:
- estrato_geográfico
- unidade_federativa
- região_administrativa
- área_tipo

A coluna estrato_geográfico serviu como chave primária da tabela dim_estratos, mantendo a ligação desta com as tabelas-fato que criamos. A unidade federativa, a região administrativa e o tipo da área foram todas incluídas nessa tabela por serem campos constantes de um mesmo estrato.

### dim_domicílios
Como dissemos, as duas entidades principais que fazem parte da PNAD são os domicílios e seus moradores. Separar os dados relativos a cada um em tabelas diferentes, por sua vez, nos dá a capacidade de realizar análises nesses dois níveis.

Os campos que utilizamos na tabela dim_domicílios são aqueles que nos permitem identificar cada domicílio ou que nos informam sobre características importantes deles, como o fato de serem próprios ou não e o valor pago pelo aluguel.

Assim, a tabela dim_domicílio agrupou as seguintes colunas:
- id_domicílio;
- peso_médio_domicílio;
- estrato_geográfico;
- domicílio_zona;
- UPA;
- domicílio_número;
- domicílio_situação;
- domicílio_valor_aluguel;
- domicílio_faixa_aluguel;
- moradores_início;
- moradores_fim

Obs. 1: a coluna peso_médio_domicílio foi calculada como a média do peso de todas as vezes em que o domicílio aparece na pesquisa.

Obs. 2: as duas últimas colunas foram criadas pela contagem do número de respostas na primeira e na última entrevistas respondidas pelo domicílio);


### Sobre as Pessoas

Cada entrada nos microdados representa um morador, sendo esta a unidade fundamental da pesquisa. Cada morador, por sua vez, pode ser unicamente identificado utilizando o id de seu domicílio + seu número de ordem. A concatenação desses valores e sua transformação em hash serviu como chave primária de todas as tabelas relacionada aos moradores que construímos. Assim, todas essas tabelas terão, como campos em comum:
- id_morador;
- id_domicílio;
- peso_médio_morador;
- UPA;
- domicílio_número;
- morador_número;

Há três principais grupos de características que identificamos para cada morador, o que resultou em três tabelas-dimensão relativas a eles.

### dim_moradores

A primeira tabela-dimensão relativa aos moradores é aquela que agrupa colunas que nos permitem identificar suas **características sociais**. Ou seja, é aquela que nos informa sobre seu sexo, raça, idade, escolaridade, etc. As colunas específicas utilizadas nesta tabela foram:
- morador_parentesco;
- morador_idade;
- morador_sexo;

### dim_saúde

O segundo grupo de características próprias de cada morador que identificamos foi o de suas **condições crônicas de saúde**: informações sobre a existência de doenças cardíacas, pulmonares, etc. As colunas específicas dessa tabela são:

### tabelas-fato

Como registram os fenômenos de interesse, as tabelas-fato geralmente são as que mais possuem colunas. Uma vez que já mencionamos quais são as 3 tabelas-fato que construímos — e uma vez que não queremos entediar o leitor com listagens enormes de colunas —, pedimos ao leitor que, caso se interesse em saber exatamente quais colunas foram incluídas em cada tabela, que procure pelas variáveis cols_fact_domicilios, cols_fact_trabalho e cols_fact_covid no código a seguir:

```python
# --- Imports ---

import sys
import json
import boto3
import geopandas as gpd
import pandas as pd

from pyspark.context import SparkContext
from pyspark.sql.types import StructType

from awsglue.utils import getResolvedOptions
from awsglue.context import GlueContext
from awsglue.job import Job

pd.set_option('display.max_columns', None)

# --- Parâmetros do Job e Inicialização do Contexto ---

args = getResolvedOptions(sys.argv, ['JOB_NAME', 'S3_BUCKET'])
bucket_name = args['S3_BUCKET']

sc = SparkContext.getOrCreate()
glueContext = GlueContext(sc)
spark = glueContext.spark_session

job = Job(glueContext)
job.init(args['JOB_NAME'], args)

s3_client = boto3.client('s3')

# --- Funções de criação de dfs ---

def create_dim_estratos(pesquisa, cols, geometry_df):
    df = pesquisa.select(*cols).toPandas()
    df = df.convert_dtypes()
    df = df.drop_duplicates()

    df = geometry_df.merge(df, on='estrato_geográfico')
    df = df[[
        'estrato_geográfico',
        'nome_estrato',
        'unidade_federativa',
        'região_administrativa',
        'area_tipo',
        'municípios',
        'geometry'
    ]].reset_index(drop=True)

    return df


def create_dim_domicilios(pesquisa, cols):
    agg_df = pesquisa.groupBy('id_domicílio', 'domicílio_peso', 'ano', 'mês').count().toPandas()
    agg_df = agg_df.sort_values(by=['id_domicílio', 'ano', 'mês'])
    agg_df = agg_df.groupby('id_domicílio', as_index=False).agg(
        peso_médio_domicílio=('domicílio_peso', 'mean'),
        moradores_início=('count', 'first'),
        moradores_fim=('count', 'last')
    ).reset_index(drop=True)

    df = pesquisa.select(*cols).toPandas()
    df = df.convert_dtypes() # Evita que colunas com NaN sejam transformadas em float.
    df = df.drop_duplicates('id_domicílio')
    df = df.merge(agg_df, on='id_domicílio').reset_index(drop=True)
    df = df[[
        'id_domicílio',
        'peso_médio_domicílio',
        'estrato_geográfico',
        'domicílio_zona',
        'UPA',
        'domicílio_número',
        'domicílio_situação',
        'domicílio_valor_aluguel',
        'domicílio_faixa_aluguel',
        'moradores_início',
        'moradores_fim'
    ]]

    return df


def create_dim_pessoas(pesquisa, cols):
    peso_medio_morador = pesquisa.select('id_morador', 'id_domicílio', 'morador_peso').toPandas()
    peso_medio_morador = peso_medio_morador.groupby('id_morador').agg(
        id_domicílio=('id_domicílio', 'first'),
        peso_médio_morador=('morador_peso', 'mean')
    ).reset_index()

    df = pesquisa.select(*cols).dropDuplicates(['id_morador']).toPandas()
    df = df.convert_dtypes()
    df = peso_medio_morador.merge(df, how='left', on='id_morador').reset_index(drop=True)

    return df

# --- Colunas dfs ---

cols_dim_estratos = [
    'estrato_geográfico',
    'unidade_federativa',
    'região_administrativa',
    'area_tipo'
]

cols_dim_domicilios = [
    'id_domicílio',
    'estrato_geográfico',
    'domicílio_zona',
    'UPA',
    'domicílio_número',
    'domicílio_situação',
    'domicílio_valor_aluguel',
    'domicílio_faixa_aluguel'
]

cols_dim_moradores = [
    'id_morador',
    'UPA',
    'domicílio_número',
    'morador_número',
    'morador_parentesco',
    'morador_idade',
    'morador_sexo',
    'morador_raça',
    'morador_escolaridade',
]

cols_dim_saude = [
    'id_morador',
    'UPA',
    'domicílio_número',
    'morador_número',
    'tem_plano_de_saúde',
    'doenças_crônicas_diabetes',
    'doenças_crônicas_hipertensão',
    'doenças_crônicas_doença_respiratória',
    'doenças_crônicas_doenças_cardíacas',
    'doenças_crônicas_depressão',
    'doenças_crônicas_câncer',
]

cols_dim_trabalho = [
    'id_morador',
    'UPA',
    'domicílio_número',
    'morador_número',
    'morador_estuda',
    'escola_pública',
    'trabalho_tem_mais_de_um',
    'trabalho_tipo',
    'trabalho_esfera',
    'trabalho_clt_ou_funcionário_público',
    'trabalho_função',
    'trabalho_ramo_empresa',
    'trabalho_tamanho_empresa',
    'trabalho_quantas_horas_trabalhava',
    'rendimento_faixa_normalmente_recebida_dinheiro',
    'rendimento_valor_normalmente_recebido_dinheiro',
    'rendimento_faixa_normalmente_recebida_produtos_e_mercadorias',
    'rendimento_valor_normalmente_recebido_produtos_e_mercadorias',
    'rendimento_recebia_somente_em_benefícios',
    'rendimento_era_não_remunerado',
    'trabalho_contribui_INSS'
]

cols_fact_domicilios = [
    'data',
    'id_domicílio',
    'domicílio_peso',
    'renda_domiciliar_aposentadoria_pensão',
    'renda_domiciliar_pensão_alimentícia_doação_mesada',
    'renda_domiciliar_bolsa_família',
    'renda_domiciliar_benefício_assistencial_continuado',
    'renda_domiciliar_auxílio_emergencial_coronavirus',
    'renda_domiciliar_seguro_desemprego',
    'renda_domiciliar_outros_rendimentos',
    'empréstimo_domicílio_alguém_pediu',
    'empréstimo_domicílio_banco',
    'empréstimo_domicílio_parente_ou_amigo',
    'empréstimo_domicílio_empregado_ou_patrão',
    'empréstimo_domicílio_outro_local_ou_pessoa',
    'profilaxia_domicílio_sabão_ou_detergente',
    'profilaxia_domicílio_álcool_70%',
    'profilaxia_domicílio_máscaras',
    'profilaxia_domicílio_luvas_descartáveis',
    'profilaxia_domicílio_água_sanitária_ou_desinfetante',
]

cols_fact_trabalho = [
    'data',
    'id_domicílio',
    'domicílio_peso',
    'id_morador',
    'morador_peso',
    'escola_aulas_presenciais',
    'escola_lição_de_casa',
    'escola_lição_de_casa_porque_não_fez',
    'escola_dedicação',
    'escola_dedicação_tempo',
    'trabalho_trabalhou_ou_fez_bico',
    'trabalho_estava_afastado',
    'trabalho_afastamento_motivo',
    'trabalho_foi_remunerado_enquanto_afastado',
    'trabalho_meses_afastado',
    'trabalho_contrato_suspenso',
    'trabalho_quantas_horas_trabalhou',
    'trabalho_queria_ter_trabalhado_mais',
    'rendimento_faixa_efetivamente_recebida_dinheiro',
    'rendimento_valor_efetivamente_recebido_dinheiro',
    'rendimento_faixa_efetivamente_recebida_produtos_e_mercadorias',
    'rendimento_valor_efetivamente_recebido_produtos_e_mercadorias',
    'trabalho_trabalhou_presencialmente',
    'trabalho_trabalhou_remotamente',
    'trabalho_procurou_emprego',
    'trabalho_porque_não_procurou_emprego',
    'trabalho_gostaria_de_ter_trabalhado',
]

cols_fact_covid = [
    'data',
    'id_domicílio',
    'domicílio_peso',
    'id_morador',
    'morador_peso',
    'sintomas_febre',
    'sintomas_tosse',
    'sintomas_dor_de_garganta',
    'sintomas_dificuldade_para_respirar',
    'sintomas_dor_de_cabeça',
    'sintomas_dor_no_peito',
    'sintomas_náusea',
    'sintomas_nariz_escorrendo',
    'sintomas_fadiga',
    'sintomas_dor_nos_olhos',
    'sintomas_perda_de_olfato',
    'sintomas_dor_muscular',
    'sintomas_diarreia',
    'providencias_procurou_atendimento',
    'providencias_ficou_em_casa',
    'providencias_ligou_para_médico',
    'providencias_automedicação',
    'providencias_tomou_remédio_orientação_médica',
    'providencias_visita_domiciliar_SUS',
    'providencias_visita_domiciliar_médico_particular',
    'providencias_outra',
    'local_atendimento_posto_de_saúde',
    'local_atendimento_pronto_socorro_SUS',
    'local_atendimento_hospital_SUS',
    'local_atendimento_consultório_privado',
    'local_atendimento_pronto_socorro_privado',
    'local_atendimento_hospital_privado',
    'internação_internou',
    'internação_entubou',
    'exames_fez_algum',
    'exames_fez_SWAB',
    'exames_resultado_SWAB',
    'exames_fez_furo_no_dedo',
    'exames_resultado_furo_no_dedo',
    'exames_fez_punção_venosa',
    'exames_resultado_punção_venosa',
    'prevenção_restringiu_contato',
]

# --- Importação dos arquivos ---

s3_client.download_file(bucket_name, 'Silver Layer/estratos_geometry.geojson', '/tmp/estratos_geometry.geojson')
geo_estratos = gpd.read_file("/tmp/estratos_geometry.geojson")

s3_client.download_file(bucket_name, 'Silver Layer/esquema_pesquisa.json', '/tmp/esquema_pesquisa.json')
with open('/tmp/esquema_pesquisa.json', 'r') as f:
    esquema = StructType.fromJson(json.load(f))

s3_silver_parquet = f"s3://{bucket_name}/Silver Layer/pesquisa_traduzida"
pesquisa = spark.read.schema(esquema).option("header", True).parquet(s3_silver_parquet)

# --- Criação das dfs ---

dim_estratos = create_dim_estratos(pesquisa, cols_dim_estratos, geo_estratos)
dim_domicilios = create_dim_domicilios(pesquisa, cols_dim_domicilios)
dim_moradores = create_dim_pessoas(pesquisa, cols_dim_moradores)
dim_saude = create_dim_pessoas(pesquisa, cols_dim_saude)
dim_trabalho = create_dim_pessoas(pesquisa, cols_dim_trabalho)
fact_domicilios = pesquisa.select(*cols_fact_domicilios)
fact_trabalho_renda = pesquisa.select(*cols_fact_trabalho)
fact_covid = pesquisa.select(*cols_fact_covid)

# --- Salvamento dos arquivos ---

dim_estratos.to_file('/tmp/dim_estratos.geojson', driver="GeoJSON")
s3_client.upload_file('/tmp/dim_estratos.geojson', bucket_name, 'Gold Layer/dim_estratos.geojson')

dim_domicilios.to_csv('/tmp/dim_domicílios.csv', index=False, sep=';')
s3_client.upload_file('/tmp/dim_domicílios.csv', bucket_name, 'Gold Layer/dim_domicílios.csv')

dim_moradores.to_csv('/tmp/dim_moradores.csv', index=False, sep=';')
s3_client.upload_file('/tmp/dim_moradores.csv', bucket_name, 'Gold Layer/dim_moradores.csv')

dim_saude.to_csv('/tmp/dim_saúde.csv', index=False, sep=';')
s3_client.upload_file('/tmp/dim_saúde.csv', bucket_name, 'Gold Layer/dim_saúde.csv')

dim_trabalho.to_csv('/tmp/dim_trabalho.csv', index=False, sep=';')
s3_client.upload_file('/tmp/dim_trabalho.csv', bucket_name, 'Gold Layer/dim_trabalho.csv')

fact_domicilios.coalesce(1).write.mode('overwrite').parquet(f's3://{bucket_name}/Gold Layer/fact_domicílios')
fact_trabalho_renda.coalesce(1).write.mode('overwrite').parquet(f's3://{bucket_name}/Gold Layer/fact_trabalho_renda')
fact_covid.coalesce(1).write.mode('overwrite').parquet(f's3://{bucket_name}/Gold Layer/fact_covid')

job.commit()
```
